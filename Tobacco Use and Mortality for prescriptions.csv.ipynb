### *Handling Missing Value*

```python
# Check missing value
prescriptions_df.isnull().sum()
```

```python
prescriptions_df.isnull().mean()
```

```python
# Fill Null Value with 0
prescriptions_df['Net Ingredient Cost of Varenicline (Champix)'] = prescriptions_df['Net Ingredient Cost of Varenicline (Champix)'].fillna(0)
```

```python
# Fill Null Value with 0
prescriptions_df['Varenicline (Champix) Prescriptions'] = prescriptions_df['Varenicline (Champix) Prescriptions'].fillna(0)
```

```python
prescriptions_df
```

### *Data Preparation, Analysis and Visualizations*

```python
# Selecting only Year in this prescriptions_df['Year'] columns
for i in range(0, len(prescriptions_df['Year'])):
    prescriptions_df['Year'][i] = (prescriptions_df['Year'][i])[:4]
```

```python
prescriptions_df['Year']
```

```python
# Check Correlation between them
plt.figure(figsize=(15,10))
sns.heatmap(prescriptions_df.corr(), annot=True)
plt.show()
```

```python
cols1 = ['Year', 'All Pharmacotherapy Prescriptions', 'Net Ingredient Cost of All Pharmacotherapies']
```

```python
# How many All Pharmacotherapy Prescriptions in year
plt.figure(figsize=(15,10))
sns.barplot(x = prescriptions_df['Year'], y = prescriptions_df['All Pharmacotherapy Prescriptions'], palette='magma')
plt.title('All Pharmacotherapy Prescriptions in year')
plt.xlabel('Year')
plt.ylabel('All Pharmacotherapy Prescriptions')
plt.show()
```
